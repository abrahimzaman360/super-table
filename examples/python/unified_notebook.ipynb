```python
import supertable as st
from pyspark.sql import SparkSession

# Load Table
table = st.open_table("s3://warehouse/events")

# 1. Polars
df_pl = table.to_polars()
print(df_pl.head())

# 2. DuckDB
import duckdb
rel = table.to_duckdb()
rel.filter("id > 100").show()

# 3. Spark
spark = SparkSession.builder \
    .config("spark.jars", "/path/to/supertable-spark.jar") \
    .getOrCreate()
    
# Native read (preferred)
df_spark = table.to_spark(spark) 
df_spark.show()
```
